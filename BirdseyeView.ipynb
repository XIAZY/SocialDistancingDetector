```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def birdseye(pts, img):
    """
    point orientation goes in order of bottom left, bottom right, top left, top right
    """
    length, width = img.shape
    
    new = np.float32([[0, length], [width, length], [0, 0], [width, 0]])
    
    M = cv2.getPerspectiveTransform(pts, new)
    
    newImg = cv2.warpPerspective(img, M, (width, length))
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(newImg, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
subway = cv2.imread("/Users/jackycen/SocialDistancingDetector/images/subway.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(10,10))
plt.imshow(subway, cmap='gray')
```

```python
pts = np.float32([[600, 3900], [2900, 3000], [50, 550], [2300, 750]])
birdseye(pts, subway)
```
